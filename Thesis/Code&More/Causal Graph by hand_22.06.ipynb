### True causal graph

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
G = nx.DiGraph()
G.add_nodes_from(['X1', 'X2', 'X3', 'X4'])
G.add_edges_from([('X1', 'X2'), ('X1', 'X3'), ('X2', 'X3'),('X3','X4')])

plt.figure(figsize=(8, 6))
#bipartite_layout(G, nodes, align='vertical', scale=1, center=None, aspect_ratio=1.3333333333333333)[source]

pos = {'X2': np.array([-1.  , -0.75]),
 'X1': np.array([-1.  ,  0.75]),
 'X3': np.array([ 1.  , 0.]),
 'X4': np.array([3 , 0])} 
nx.draw_networkx(G, pos, with_labels=True, node_size=1000, node_color='lightgrey', font_size=12, font_weight='bold', arrows=True)
plt.title('Simpon')
plt.axis('off')  # Disable axis
plt.show()

```

### True SCM

```python
import matplotlib.pyplot as plt

def Or_Causal():
    # Set up the figure and axis
    #fig, ax = plt.subplots()

    # Set the equation as a LaTeX expression
    equation1 = r'$x_1 : f_1= u_1$'
    equation2 = r'$X_2 : f_2(x_1,u_2) = a_1x_1 +u_2$'
    equation3 = r'$X_3 : f_3(x_1,x_2,u_3) = b_1x_1 +b_2x_2+u_3$'
    equation4 = r'$X_4 : f_4(x_3,u_4) = c_1x_3 +u_4$'
    
    equation = equation1+'\n'+equation2+'\n'+equation3+'\n'+ equation4
    # Plot the equation as text
    #ax.text(0.5, 0.5, equation, fontsize=18, ha='center')

    # Hide the axes and labels
    #ax.axis('off')

    # Show the plot
    #plt.show()
    return equation
```

```python
Scm = Or_Causal()
```

```python
flowchart = nx.DiGraph()
```

```python
Scm = Or_Causal()

flowchart.add_edge('Gausian \n distribution',Scm)
flowchart.add_edge(Scm, 'Data \n generation')

```

```python
pos = {'Gausian \n distribution': np.array([-1, 0   ]),
 '$x_1 : f_1= u_1$\n$X_2 : f_2(x_1,u_2) = a_1x_1 +u_2$\n$X_3 : f_3(x_1,x_2,u_3) = b_1x_1 +b_2x_2+u_3$\n$X_4 : f_4(x_3,u_4) = c_1x_3 +u_4$': np.array([ 0    , 0]),
 'Data \n generation': np.array([1,  0   ]),
 }
```

```python

nx.draw(flowchart, pos, with_labels=True, node_size=10000, node_color='lightgrey', font_size=12, font_weight='bold', edge_color='black', arrows=True)
plt.show()
```

#### Data generation

If there is no intervention, based on the SCM above,we can get the first data point:$\newline$
let $u_1 = 0,5$,$u_2 = 0.3$,$u_3 = 0.7$,$u_4=0.1$ $\newline$ $a_1 =1$,$b_1 = 2$,$b_2 = 3$,$c_1 = 4$ 

The equations are $X_1 : f_1= u_1 \\ X_2 : f_2(x_1,u_2) = x_1 +u_2 \\ X_3 : f_3(x_1,x_2,u_3) = 2x_1 +3x_2+u_3 \\
   X_4 : f_4(x_3,u_4) = 4x_3 +u_4$ $\newline$,the data point is $x^f=$ 
$[x_1 = 0.5 \newline x_2 = 0.8 \newline x_3 =4.1 \newline x_4 = 16.5]$


### How the intervention effect the counterfactors

### If we do intervention on $x_1$,which is $do(x_1 = 0)$

, then the **true causal graph**changes to 

```python
G = nx.DiGraph()
G.add_nodes_from([r'$X_1=0$', r'$X_2$', r'$X_3$', r'$X_4$'])
G.add_edges_from([(r'$X_2$', r'$X_3$'),(r'$X_3$', r'$X_4$')])

plt.figure(figsize=(8, 6))
#bipartite_layout(G, nodes, align='vertical', scale=1, center=None, aspect_ratio=1.3333333333333333)[source]

pos = {r'$X_2$': np.array([-1.  , -0.75]),
r'$X_1=0$': np.array([-1.  ,  0.75]),
 r'$X_3$': np.array([ 1.  , 0.]),
 r'$X_4$': np.array([3 , 0])} 
nx.draw_networkx(G, pos, with_labels=True, node_size=1000, node_color='lightgrey', font_size=12, font_weight='bold', arrows=True)
plt.title('Simpon')
plt.axis('off')  # Disable axis
plt.show()
```

### True couterfactual point

Then we get the **true couterfactual point** $x^{cf}=$ 
$[x_1 = 0 \newline x_2 = 0.3 \newline x_3 =1.6 \newline x_4 = 6.5]$


### Remove

Suppose we get a wrong causal graph, where the edge between $x_2$ and $x_3$ is **removed**, the causal graph looks like 

```python
G = nx.DiGraph()
G.add_nodes_from([r'$X_1$', r'$X_2$', r'$X_3$', r'$X_4$'])
G.add_edges_from([(r'$X_1$', r'$X_2$'),(r'$X_1$', r'$X_3$'),(r'$X_3$', r'$X_4$')])

plt.figure(figsize=(8, 6))
#bipartite_layout(G, nodes, align='vertical', scale=1, center=None, aspect_ratio=1.3333333333333333)[source]

pos = {r'$X_2$': np.array([-1.  , -0.75]),
r'$X_1$': np.array([-1.  ,  0.75]),
 r'$X_3$': np.array([ 1.  , 0.]),
 r'$X_4$': np.array([3 , 0])} 
nx.draw_networkx(G, pos, with_labels=True, node_size=1000, node_color='lightgrey', font_size=12, font_weight='bold', arrows=True)
plt.title('Simpon')
plt.axis('off')  # Disable axis
plt.show()
```

The corresponding equations are $X_1 : f_1= \tilde{u_1} \\ X_2 : f_2(x_1,u_2) = x_1 +\tilde{u_2} \\ X_3 : f_3(x_1,u_3) = 2x_1 + \tilde{ u_3 }\\
   X_4 : f_4(x_3,u_4) = 4x_3 +\tilde{u_4}$ 

```python
def exogenous_variables(x1,x2,x3,x4):
    
    u_1 = round((x1),2)
    
    u_2 = round((x2 - x1),2)
    
    u_3 = round((x3 - 2*x1),2)
     
    u_4 = round((x4 - 4*x3),2)
    
    return [u_1, u_2, u_3, u_4]
```

```python
remove_exo = exogenous_variables(0.5,0.8,4.1,16.5)
remove_exo
```

The given data is $x^f=$ 
$[x_1 = 0.5 \newline x_2 = 0.8 \newline x_3 =4.1 \newline x_4 = 16.5]$, then we get the exogenous variables values as:
$\tilde{u_1} = 0.5, \space \tilde{u_2} = 0.3 ,\space \tilde{u_3} = 3.1, \space \tilde{u_4} = 0.1$



If we do intervention, the causal graph will be 

```python
G = nx.DiGraph()
G.add_nodes_from([r'$X_1=0$', r'$X_2$', r'$X_3$', r'$X_4$'])
G.add_edges_from([(r'$X_3$', r'$X_4$')])

plt.figure(figsize=(8, 6))
#bipartite_layout(G, nodes, align='vertical', scale=1, center=None, aspect_ratio=1.3333333333333333)[source]

pos = {r'$X_2$': np.array([-1.  , -0.75]),
r'$X_1=0$': np.array([-1.  ,  0.75]),
 r'$X_3$': np.array([ 1.  , 0.]),
 r'$X_4$': np.array([3 , 0])} 
nx.draw_networkx(G, pos, with_labels=True, node_size=1000, node_color='lightgrey', font_size=12, font_weight='bold', arrows=True)
plt.title('Simpon')
plt.axis('off')  # Disable axis
plt.show()
```

```python
def counterfactual(x1,x2,x3,x4,u1,u2,u3,u4):
    x1 =0
    x2 = u2
    x3 = u3
    x4 = 4*x3 + u4
    
    return [x1,x2,x3,x4]
```

```python
remove_x1_cf = counterfactual(0.5,0.8,4.1,16.5,0.5,0.3,3.1,0.1)
remove_x1_cf
```

### Remove the edge from x2 to x3,the counterfactual about do(x1=0)

We get the point $x^{cf1} = [0,\newline 0.3,\newline 3.1,\newline 12.5]$

### Change Direction

Suppose we get a wrong causal graph, where the direction of edge between $x_2$ and $x_3$ is **changed**, the causal graph looks like 

```python
G = nx.DiGraph()
G.add_nodes_from([r'$X_1$', r'$X_2$', r'$X_3$', r'$X_4$'])
G.add_edges_from([(r'$X_1$', r'$X_2$'),(r'$X_3$', r'$X_2$'),(r'$X_1$', r'$X_3$'),(r'$X_3$', r'$X_4$')])

plt.figure(figsize=(8, 6))

pos = {r'$X_2$': np.array([-1.  , -0.75]),
r'$X_1$': np.array([-1.  ,  0.75]),
 r'$X_3$': np.array([ 1.  , 0.]),
 r'$X_4$': np.array([3 , 0])} 
nx.draw_networkx(G, pos, with_labels=True, node_size=1000, node_color='lightgrey', font_size=12, font_weight='bold', arrows=True)
plt.title('Simpon')
plt.axis('off')  # Disable axis
plt.show()
```

The corresponding equations are $X_1 : f_1= \tilde{u_1} \\ X_2 : f_2(x_1,x_3,u_2) = x_1 +3x_3 + \tilde{u_2} \\ X_3 : f_3(x_1,u_3) = 2x_1 + \tilde{ u_3 }\\
   X_4 : f_4(x_3,u_4) = 4x_3 +\tilde{u_4}$ 

```python
def exogenous_variables2(x1,x2,x3,x4):
    
    u_1 = round((x1),2)
    
    u_2 = round((x2 - x1 - 3*x3),2)
    
    u_3 = round((x3 - 2*x1),2)
     
    u_4 = round((x4 - 4*x3),2)
    
    return [u_1, u_2, u_3, u_4]
```

```python
change_exo = exogenous_variables2(0.5,0.8,4.1,16.5)
change_exo
```

The given data is $x^f=$ 
$[x_1 = 0.5 \newline x_2 = 0.8 \newline x_3 =4.1 \newline x_4 = 16.5]$, then we get the exogenous variables values as:
$\tilde{u_1} = 0.5, \space \tilde{u_2} = -12.0 ,\space \tilde{u_3} = 3.1, \space \tilde{u_4} = 0.1$

**Intervention**

```python
G = nx.DiGraph()
G.add_nodes_from([r'$X_1=0$', r'$X_2$', r'$X_3$', r'$X_4$'])
G.add_edges_from([(r'$X_3$', r'$X_4$'),(r'$X_3$', r'$X_2$')])

plt.figure(figsize=(8, 6))
#bipartite_layout(G, nodes, align='vertical', scale=1, center=None, aspect_ratio=1.3333333333333333)[source]

pos = {r'$X_2$': np.array([-1.  , -0.75]),
r'$X_1=0$': np.array([-1.  ,  0.75]),
 r'$X_3$': np.array([ 1.  , 0.]),
 r'$X_4$': np.array([3 , 0])} 
nx.draw_networkx(G, pos, with_labels=True, node_size=1000, node_color='lightgrey', font_size=12, font_weight='bold', arrows=True)
plt.title('Simpon')
plt.axis('off')  # Disable axis
plt.show()
```

```python
def counterfactual2(x1,x2,x3,x4,u1,u2,u3,u4):
    x1 =0
    x2 = 3*x3 + u2
    x3 = u3
    x4 = 4*x3 + u4
    
    return [x1,x2,x3,x4]
```

```python
change_x1_cf = counterfactual2(0.5,0.8,4.1,16.5,0.5,-12.0,3.1,0.1)
change_x1_cf
```

We get the point $x^{cf2} = [0,\newline 0.3,\newline 3.1,\newline 12.5]$

### Add some edge

```python
G = nx.DiGraph()
G.add_nodes_from(['X1', 'X2', 'X3', 'X4'])
G.add_edges_from([('X1', 'X2'), ('X1', 'X3'), ('X2', 'X3'),('X2','X4'),('X3','X4')])

plt.figure(figsize=(8, 6))
#bipartite_layout(G, nodes, align='vertical', scale=1, center=None, aspect_ratio=1.3333333333333333)[source]

pos = {'X2': np.array([-1.  , -0.75]),
 'X1': np.array([-1.  ,  0.75]),
 'X3': np.array([ 1.  , 0.]),
 'X4': np.array([3 , 0])} 
nx.draw_networkx(G, pos, with_labels=True, node_size=1000, node_color='lightgrey', font_size=12, font_weight='bold', arrows=True)
plt.title('Simpon')
plt.axis('off')  # Disable axis
plt.show()
```

The corresponding equations are $X_1 : f_1= \tilde{u_1} \\ X_2 : f_2(x_1,u_2) = x_1 +\tilde{u_2} \\ X_3 : f_3(x_1,u_3) = 2x_1 + 3x_2+\tilde{ u_3 }\\
   X_4 : f_4(x_3,u_4) = 4x_3 +5x_2+\tilde{u_4}$ 

```python
def exogenous_variables3(x1,x2,x3,x4):
    
    u_1 = round((x1),2)
    
    u_2 = round((x2 - x1),2)
    
    u_3 = round((x3 - 2*x1 - 3*x2),2)
     
    u_4 = round((x4 - 4*x3 - 5*x2),2)
    
    return [u_1, u_2, u_3, u_4]
```

```python
add_exo = exogenous_variables3(0.5,0.8,4.1,16.5)
add_exo
```

The given data is $x^f=$ 
$[x_1 = 0.5 \newline x_2 = 0.8 \newline x_3 =4.1 \newline x_4 = 16.5]$, then we get the exogenous variables values as:
$\tilde{u_1} = 0.5, \space \tilde{u_2} = 0.3 ,\space \tilde{u_3} = 0.7, \space \tilde{u_4} = -3.9$

**Intervention**

```python
G = nx.DiGraph()
G.add_nodes_from([r'$X_1=0$', r'$X_2$', r'$X_3$', r'$X_4$'])
G.add_edges_from([(r'$X_3$', r'$X_4$'),(r'$X_2$', r'$X_3$'),(r'$X_2$', r'$X_4$')])

plt.figure(figsize=(8, 6))
#bipartite_layout(G, nodes, align='vertical', scale=1, center=None, aspect_ratio=1.3333333333333333)[source]

pos = {r'$X_2$': np.array([-1.  , -0.75]),
r'$X_1=0$': np.array([-1.  ,  0.75]),
 r'$X_3$': np.array([ 1.  , 0.]),
 r'$X_4$': np.array([3 , 0])} 
nx.draw_networkx(G, pos, with_labels=True, node_size=1000, node_color='lightgrey', font_size=12, font_weight='bold', arrows=True)
plt.title('Simpon')
plt.axis('off')  # Disable axis
plt.show()
```

```python
def counterfactual3(x1,x2,x3,x4,u1,u2,u3,u4):
    x1 =0
    x2 = u2
    x3 = 4*x2 + u3
    x4 = 4*x3 + 5*x2 + u4
    
    return [x1,x2,x3,x4]
```
